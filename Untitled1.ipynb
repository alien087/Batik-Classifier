```python
from PIL import Image
from PIL import ImageEnhance
```

```python
a = Image.open('cap.jpg')
```

```python
a
```

```python
enhancer = ImageEnhance.Sharpness(a)
```

```python
for i in range(8):
    factor = i / 4.0
    enhancer.enhance(factor).show("Sharpness %f" % factor)
```
